```python
import sqlalchemy
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float, Date
from sqlalchemy import * 
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session

import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

from flask import Flask, jsonify

from sqlalchemy.ext.declarative import declarative_base
```

```python
#dates 2016-06-12 thru 2016-06-22
engine = create_engine('sqlite:///Resources/database.sqlite')
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
# Save reference to the table
Measurements = Base.classes.measurements
Stations = Base.classes.stations
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)

sel = [Measurements.date, Measurements.prcp]
year_query = (session
             .query(* sel)
             .filter(and_(Measurements.date >= '2015-06-12', Measurements.date <= '2016-06-12')))

df = pd.read_sql_query(year_query.statement, year_query.session.bind)

df.set_index('date', inplace=True)
daily_average = df.groupby('date').mean()
daily_average.plot.line()
plt.tight_layout()
plt.show()

#PUT IN ANGLES and UP/DOWN orientation for dates
```

```python
#Design a query to calculate the total number of stations.
station_count = (session
                .query(Measurements, func.count().label('measurement_count'))
                .group_by(Measurements.station)
                .order_by('measurement_count DESC'))
station_count.all()

df2 = pd.read_sql_query(station_count.statement, station_count.session.bind)
print("The station with the highest number of observations is " + (station_count.first()[0].station) + ".")
```

```python
tobs_query = (session
             .query(Measurements.tobs)
             .filter(and_(Measurements.date >= '2015-06-12', Measurements.date <= '2016-06-12'))
             .filter(Measurements.station == 'USC00519281'))

df3 = pd.read_sql_query(tobs_query.statement, tobs_query.session.bind)

df3.hist(bins=12)
plt.tight_layout()

plt.show()

# plt.xlabel('Temperatures')
# plt.ylabel('Number of Observations')
# plt.title('Histogram of Temperature Observance Frequency at Station USC00519281')
```

```python
#Write a function called `calc_temps` that will accept a start date and end date in the format `%Y-%m-%d` and return the minimum, average, and maximum temperatures for that range of dates.
def calc_temps(start_date, end_date):
    temps = (session
            .query(Measurements.tobs,
                  func.min(Measurements.tobs).label('min_temp'),
                  func.max(Measurements.tobs).label('max_temp'),
                  func.avg(Measurements.tobs).label('avg_temp'))
            .filter(and_(Measurements.date >= start_date, Measurements.date <= end_date))
            .filter(Measurements.station == 'USC00519281'))
    min = temps.one().min_temp
    max = temps.one().max_temp
    avg = temps.one().avg_temp
    print("Min. Temp: " + str(min))
    print("Max. Temp: " + str(max))
    print("Avg. Temp: " + str(avg))
    return[min, max, avg]
    
min, max, avg = calc_temps('2015-06-12','2016-06-12')

# * Plot the min, avg, and max temperature from your previous query as a bar chart.
#   * Use the average temperature as the bar height.
#   * Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr).
plt.bar(x = 0, height=avg, yerr=(max-min)/2)
plt.tight_layout()
plt.show()

#Use the `calc_temps` function to calculate the min, avg, and max temperatures for your trip using the matching dates from the previous year
```
